```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()


df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country']=='United States'].copy()
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
```

```python
df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

df_US_pivot
```

```python
df_US_pivot.sort_values('month_no', inplace=True)

df_US_pivot.set_index('job_posted_month', inplace = True)
```

```python
df_US_pivot.drop('month_no', axis=1, inplace=True)
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_5 = df_US['job_title_short'].value_counts().head(5)

top_5 = top_5.index.to_list()
```

```python
df_US_pivot[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')

df_US_software_pivot
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
top_5_2 = df_US_merged.sum().sort_values(ascending=False).head().index.to_list()

top_5_2
```

```python
#plt.figure(figsize=(10,6))
df_US_merged[top_5_2].plot(kind='line', figsize=(10,6))
plt.title('Monthly Job Postings for Top 5 Tech Jobs')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
